2. Python excercise: 
    Print integers 1 to N, but print "Fizz" if an integer is divisible by 3, "Buzz" if an integer is divisible by 5, and "FizzBuzz" if an integer is divisible by both 3 and 5.

```python
for number in range(1,50):
    if number % 3 is 0:
        print("Fizz")
    elif number % 5 is 0:
        print("Buzz")
    elif number % 5 is 0:
        print("FizzBuzz")
    else:
        print(number)
```
